```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
#Read the dataset
df=pd.read_csv('/content/Telco_Customer_Churn_Dataset .csv')
df
```

```python
#Describe to the dataset
df.describe()
```

```python
#Drop the CustomerId Column because its useless
df = df.drop(columns=['customerID'])
df
```

```python
#First Count the string value in TotalCharge Course
count=0
for i in df.TotalCharges:
    if i==' ':
        count+=1
print('count of empty string:- ',count)
#Replace blank value with nan value
df['TotalCharges'] = df['TotalCharges'].replace(" ",np.nan)
#Convert Object Datatype into float datatype
df['TotalCharges'] = df['TotalCharges'].astype(float)

```

```python
#Check the datatype of TotalCharges
df.info()
```

```python
#Find the missing value present or not in columns
df.isnull().sum()
```

```python
#Fill the missing value with mean value of totalcharges column
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
#Again check the missing value is filled or not
df.isnull().sum()
```

```python
#numerical variables

num = list(df.select_dtypes(include=['int64','float64']).keys())

#categorical variables

cat = list(df.select_dtypes(include='O').keys())

print(cat)
print(num)
```

```python
#Counting the value of each categorial Column
# for i in cat:
#     print(df[i].value_counts())
#There is one category which need to convert into No category
df.MultipleLines = df.MultipleLines.replace('No phone service','No')
df.OnlineSecurity = df.OnlineSecurity.replace('No internet service','No')
df.OnlineBackup = df.OnlineBackup.replace('No internet service','No')
df.DeviceProtection = df.DeviceProtection.replace('No internet service','No')
df.TechSupport = df.TechSupport.replace('No internet service','No')
df.StreamingTV = df.StreamingTV.replace('No internet service','No')
df.StreamingMovies = df.StreamingMovies.replace('No internet service','No')
```

```python
# We need to convert all category value like Yes to 1 and No to 0
for i in cat:
    df[i] = df[i].replace('Yes',1)
    df[i] = df[i].replace('No',0)
```

```python
#Check the value is converted properly or not
df
```

```python
# Convert Gender columns value like Yes into 1 and No into 0
df.gender = df.gender.replace('Male',1)
df.gender = df.gender.replace('Female',0)
```

```python
#First we need to convert category data from object to string
# Then Convert String Category values into numerical value like 0 1 2 3
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
columns_to_encode = ['InternetService', 'Contract', 'PaymentMethod']
for column in columns_to_encode:
    df[column] = df[column].astype(str)
for column in columns_to_encode:
    df[column] = label.fit_transform(df[column])
```

```python
# Convert Big numerical value into smaller numerical value
scale_cols = ['tenure','MonthlyCharges','TotalCharges']
# Using MinMaxScaler easily convert into smaller numrerical value
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
df[scale_cols] = scale.fit_transform(df[scale_cols])
```
